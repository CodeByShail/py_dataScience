```python
# Accessing Elements from list--you can get the elements with the help of indexing
fruits = ["Apple","Banana","kiwi"]
print(fruits)
print(fruits[1])
print(fruits[2])
print(fruits[-2])
```

```python
# modifying list
fruits = ["Apple","Banana","kiwi"]
print(fruits)
fruits[1]= "orange"
print(fruits)
fruits[2]= 678
print(fruits)
```

```python
# List operations
# append---add any itemat lat(end)
# insert---add any item at particular/specific index
# remove---remove/delete item by its value
# pop ---remove last element of list

# append operation
fruits = ["Apple","Banana","kiwi"]
print(fruits)
fruits.append("orange")
print(fruits)
      
```

```python
# insert operation
fruits = ["Apple","Banana","kiwi"]
print(fruits)
fruits.insert(1,"orange")
print(fruits)
    
```

```python
# remove
fruits = ["Apple","Banana","kiwi"]
print(fruits)
fruits.remove("Banana")
print(fruits)
```

```python
# pop
fruits = ["Apple","Banana","kiwi"]
print(fruits)
popped=fruits.pop()
print(popped)
print(fruits)
```

```python
# list silicing--[startindex:endIndex(exclusive)]
num = [10,20,30,40,50]
print(num)
print(num[1:4])
print(num[:4])
print(num[1:])
print(num[:])
```

```python
# built-in functions
# len()---this function gives the length/count of all elements in list
# max()--this function gives the maximun values
# min()--this function gives the minimun values
# sort()--this function gives the sorted list(ascending order--small-large)

num = [100,455,10,20,80,50]
print(num)
print(len(num))
print(max(num))
print(min(num))
num.reverse()
print(num)
num.sort()
print(num)
# num.reverse()
# print(num)
# sort function parameter
num.sort(reverse=True)
print(num)
```

```python
# Tuple--Tuple is ordered ,immutable(inchangeble) collection of items

# syntax for creating tuple
# tupleNMae = (item1,item2..)
t1=(1,3,5,4)
print(t1)
```

```python
# Accessing elements from tuple
t1=(1,3,5,4)
print(t1)
print(t1[3])
```

```python
# imMutability 
t1= (10,20,30)
print(t1)
# t1[2] = 47 //Error(immutability-unchanged)
print(t1)
```

```python
# Tuple operations-
# len--lenghth of tuple
# max--maximum value of tuple
# min--minimum value of tuple
# sum--sum adds all the elements of tuple

t1 = (1,2,3)
print(t1)
print(t1[1]+t1[2])
print(t1[1]-t1[2])
print(t1[1]*t1[2])
print(len(t1))
print(max(t1))
print(min(t1))
print(sum(t1))
```

```python
# tuple packing and unpacking
person = ("shail",19,"Engineer") #packing
print(person)

name = person[0] 
print(name)
age = person[1]
print(age)
# unpacking
nam,age,job=person
print(name)
print(age)
```

```python
# singleton Tuple--A tuple with only 1 value in it is called singleton tuple(with trailing comma)
t1=(1)
print(type(t1))
t1 = 1
print(type(t1))
t1 = (1,)
print(type(t1))
t2 = 1,2,3
print(type(t2))
print(t2)
```

```python
""" 1. WAP to create a list of 5 fruits and perform following operations:
    -> replace one element
    ->Add a new element
    ->remove one element
    ->print the final list
"""

fruits = ["mango","Apple","orange","banana","kiwi"]
print(fruits)
# Replace
fruits[2] = "bus"
print(fruits)

# add new element
fruits.append("orange")
print(fruits)

# remove one element
fruits.remove("bus")
print(fruits)

# final list
print(fruits)
```

```python
# 2. WAP to create tuple of your name,age and city 
#       ->use unpacking to print them separately

t1 = ("shail",19,"sitamarhi")
print(t1)
# unpacking
name,age,city = t1
print(name)
print(age)
print(city)
```
