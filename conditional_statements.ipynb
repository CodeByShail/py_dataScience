```python
'''
Que 1.create a program that determines whether a given year is a leap year or not. A leap year is divisible by 4, but if it is divisible by 100, it must also be divisible by 400.
'''

year = int (input("enter year:"))
if year % 4 ==0 and year %100 !=0 or year %400 ==0:
    print(f" {year} is leap year")
else:
    print(f" {year} is  not leap year")
```

```python
#Que 2. Implement a simple calculator program that takes two numbers and an operator(+,-,*,/) as input and performs the corresponding operations.

num1 = int(input("enter first number:"))
num2 = int(input("enter second number:")) 
op = input("enter operator(+,-,*,/):")
if op =='+':
    print("sum:",num1+num2)
elif op == '-':
    print("substraction:",num1-num2)
elif op == '*':
    print("multiplication:",num1*num2)
elif op =='/':
    print("division:",num1/num2)
else :
  print("invalid ")
```

```python
#Ques 3. Implement a simple calculator program that takes two numbers and an operator(+,-,*,/) as input and performs the corresponding operations using macing case.

a = int(input("enter first number:"))
b = int(input("enter second number:")) 
op = input("enter operator(+,-,*,/):")

match op:
    case '+':
        result = a + b
    case '-':
        result = a - b
    case '*':
        result = a*b
    case '/':
        if (b !=0):
            result = a/b
        else:
            result = "Error : Divided by zero:"
    case _:
        result = "Error : Invalid input:"
print(f"The result is {result}")
```
