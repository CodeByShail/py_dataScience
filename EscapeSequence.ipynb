```python
# # escape sequence
# \n--new line character
# \r --carriage return
# \t--tab space
# \b--back space
# \'-- use to print single quote
# \"--use to print double quotes
# \\--use to print back slash
```

```python
print("Hello world :")
print('hello world:') 

print("\Hello \world:")
print("hello \nworld") 
print("\hello \\n \world")# espace sequence
print("hello \r world") 
print("hello \nworld") 
print("Hello \bworld")
print("hello \tworld") 
print("hello \' world") 
print("Hello \" world")
print("hello \\ world") 
```
