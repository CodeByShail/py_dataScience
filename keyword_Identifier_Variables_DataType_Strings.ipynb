```python
# string silicing--to access a particular part of a string and we can access that part of string with in range
# syntax--variableName[startIndex(incluxive):endIndex(exclusive)]
text = "programming"
print(text[3:7])
print(text[0:7])
print(text[ :7])
print(text[: ])
print(text[ :-1])
```
